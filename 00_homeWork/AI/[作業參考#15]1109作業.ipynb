# 熱力圖(Seaborn-Heatmap)


![](https://miro.medium.com/v2/resize:fit:720/format:webp/1*h3UCYiFCw69hL5ljIbR--A.png)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 簡單熱力圖

使用np函數設定隨機種子，同樣也是使用np函數，從均勻分配[0~1）之間抽出2*3陣列。要特別注意，這裡抽取的範圍包含0，但不包含1。接下來使用heatmap函數繪製出熱力圖，同時在最後一行執行uniform_data，可以顯示出隨機抽取2*3的數值。

```python
# 設定隨機種子
np.random.seed(0)

# 從均勻分配[0~1)之間抽出2*3陣列
uniform_data = np.random.rand(2,3)

sns.heatmap(uniform_data)

uniform_data
```

```python
# 使用10*10均勻分配資料
np.random.seed(0)

uniform_data = np.random.rand(10,10)

sns.heatmap(uniform_data)
```

## 顏色範圍

一樣使用前面10*10均勻分配資料，數值範圍從0.5~0.8，這代表右邊長條的範圍，同樣是數值愈大則顏色愈淺。

```python
# 設定顏色範圍
sns.heatmap(uniform_data, vmin=0.5, vmax=0.8)
```

## 設定中心點

可以使用center參數來設定中心點，這裡將顏色的中心點設為0.5。可以看到圖中右邊的數據，中心點0.5為黑色，0.5以下的地方開始逐漸轉為藍色，而0.5以上開始逐漸轉為乳白色。

```python
# 設定中心點
sns.heatmap(uniform_data, center=0.5)
```

```python
# 抽取標準常態分配10*10數據,以零為中心點
np.random.seed(0)

normal_data = np.random.randn(10, 10)

sns.heatmap(normal_data, center=0)
```

### 常態分配圖

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# 生成一組隨機數據，符合常態分佈
mean = 0  # 平均值
std_dev = 1  # 標準差
num_samples = 1000  # 數據點的數量

data = np.random.normal(mean, std_dev, num_samples)

# 計算常態分佈的概率密度函數
x = np.linspace(mean - 3*std_dev, mean + 3*std_dev, 1000)
pdf = norm.pdf(x, mean, std_dev)

# 繪製常態分佈曲線
plt.plot(x, pdf, color='black')
plt.title('68-95-99.7 Rule')

plt.ylabel('Probability Density')

# 設定 x 軸刻度
plt.xticks([mean - 3*std_dev, mean - 2*std_dev, mean - std_dev, mean, mean + std_dev, mean + 2*std_dev, mean + 3*std_dev],
           [r'$\mu-3\alpha$', r'$\mu-2\alpha$', r'$\mu-\alpha$', r'$\mu$', r'$\mu+\alpha$', r'$\mu+2\alpha$', r'$\mu+3\alpha$'])

# 標記68%, 95%, 99.7%區間的顏色
plt.fill_between(x, pdf, where=(x >= mean-std_dev) & (x <= mean+std_dev), alpha=0.5, color='red', label='68%')
plt.fill_between(x, pdf, where=(x >= mean-2*std_dev) & (x <= mean+2*std_dev) & (x < mean-std_dev), alpha=0.5, color='blue', label='95%')
plt.fill_between(x, pdf, where=(x >= mean-2*std_dev) & (x <= mean+2*std_dev) & (x >= mean+std_dev), alpha=0.5, color='blue')
plt.fill_between(x, pdf, where=(x >= mean-3*std_dev) & (x <= mean-2*std_dev), alpha=0.5, color='green', label='99.7%')
plt.fill_between(x, pdf, where=(x >= mean+2*std_dev) & (x <= mean+3*std_dev), alpha=0.5, color='green')


plt.legend()

plt.show()
```

## 顏色

使用cmap參數可以設定顏色，第一個範例為均勻分配，顏色使用綠色，需注意綠色第一個英文字要大寫，且後面要加上英文的複數s。

使用xticklabels及yticklabels參數可以設定刻度。xticklabels設定間隔為2，yticklabels設定間隔為3。

## 正方圖

先繪製出原始圖形做為比較的基礎，設定隨機種子並從均勻分配[0~1）之間抽出3*3陣列。

```python
# 設定隨機種子
np.random.seed(0)

# 從均勻分配[0~1)之間抽出3*3陣列
uniform_data = np.random.rand(3,3)

sns.heatmap(uniform_data)
```

## 飛行數據

首先戴入飛行數據並檢視前五筆資料，可以看到第一個欄位為年，第二個欄位為月，最後一個欄位為當月的載客人數。

```python
# 載入飛行數據
flights = sns.load_dataset('flights')
flights.head()
```

```python
# 飛行數據統計資料
flights.describe()
```

```python
# 使用pivot函數將資料轉為DataFrame結構
flights_df = flights.pivot('month', 'year', 'passengers')
flights_df
```

```python
# 繪製飛行數據熱力圖
sns.heatmap(flights_df)
```

顯示圖中數據

範例一，取到整數，在這裡將參數annot設為True代表在圖中顯示出數值，而fmt=’d’代表取到整數位。

```python
# 範例一，取到整數
sns.heatmap(flights_df, annot=True, fmt='d')
```

```python
# 範例二，取到小數點第一位
sns.heatmap(flights_df, annot=True, fmt='.1f')
```

```python
# 放大圖形
plt.figure(figsize=(8,8))
sns.heatmap(flights_df, annot=True, fmt='.1f')
```

```python
# 範例三,常態分配取到小數點第一位
sns.heatmap(normal_data, annot=True, fmt='.1f')
```

## 線條設定
範例一，線條寬度選擇0.5，預設顏色為白色。可以看到圖中每個方格之間都用白色做為間隔。

```python
# 範例二，線條寬度選擇1.5與0.5做比較
sns.heatmap(flights_df, linewidths=1.5)
```

```python
# 範例三，線條顏色選黑色
sns.heatmap(flights_df, linewidths=0.5, linecolor='black')
```

```python
# 範例四，線條顏色選青色
sns.heatmap(flights_df, linewidths=0.5, linecolor='cyan')
```
