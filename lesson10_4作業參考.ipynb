擲骰子遊戲
1.至少要有兩個骰子點數相同，另兩個骰子才加總計算點數。

-例:[1,2,3,4]->0點 [1,1,2,3]->5點

2.兩兩相同以點數高者計算。

-例:[1,1,6,6]->12點

3.4個骰子相同，點數乘10倍。

-例:[1,1,1,1]->40點

4.其餘不計算點數。

5.與電腦點數相同者電腦羸。

```python
#林玫蓀
import random
import pyinputplus as pyip

def diceGame():    
    j = 0
    t = 0
    score_t = 0 

    num = [random.randint(1,6) for i in range(4) ]  
    
    print(num,end=" ")
    for x in range(4):
        for j in range(x+1,4) :
            if num[x] == num[j] :
                t += 1
                score_t += num[j]            
    if t == 1 :
        point = sum(num)-score_t*2
        return point
    elif  t == 2:
        point = int(sum(num)/2-min(num))
        return point
    elif  t == 6:
        point = int(sum(num)*10)
        return point
    else :
        point = 0
        return point 


while True:
    print()
    print("===============擲骰子遊戲=================:\n")     
    a = diceGame()
    print(f"電腦擲骰子點數:{a}點")    
    b = diceGame()    
    print(f"您擲骰子點數:{b}點")
    if b > a:
        print("you win") 
    else :
        print("you lose")
    answer = pyip.inputYesNo(prompt="還要繼續嗎?(y,n)",yesVal='y',noVal='n')
    if not (answer == 'y'):
        print()
        break
print("遊戲結束")
```
